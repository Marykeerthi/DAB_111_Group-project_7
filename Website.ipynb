```python
with open ("Data.csv") as file:
    data_file=csv.reader(file)
    for row in data_file:
        if row[0]=='EMPLOYEEID':
            continue
        cursor.execute("INSERT INTO Salary_of_professionals VALUES(?,?,?,?,?,?,?,?)",(row[0],row[1],row[2],row[3],row[4],row[5],row[6],row[7]))
```
